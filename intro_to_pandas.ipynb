```python
# Importing a database
sales_2_df = pd.read_excel('Vendas.xlsx') # If the file is in a different folder write down the path
display(sales_2_df)
```

```python
# EDITING A DATAFRAME 
# getting a single column from a df (pd.Series)
products = sales_2_df['Produto'] #products is not a df but a series
products
```

```python
# getting more than a column from a df (pd.Series)
products_2 = sales_2_df[['Produto', 'ID Loja']]
products_2 # more than 1 series = table
```

```python
# method .loc (VERY IMPORTANT)
# Getting a single row
sales_2_df.loc[1] # a row is also a series
```

```python
# Getting more than a row
sales_2_df.loc[1: 5]
```

```python
# Getting lines which correspond a condition
# SYNTAX => df.loc[CONDITION]

# Storing it in a df
sales_norteshopping_df = sales_2_df.loc[sales_2_df['ID Loja'] == 'Norte Shopping']
sales_norteshopping_df

```

```python
# Selecting rows and colmuns with df.loc[rows, cols]

sales_norteshoppping_2_df = sales_2_df.loc[sales_2_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
sales_norteshoppping_2_df                                   
```

```python
# Picking up a specific value with the methot df.loc[]

sales_2_df.loc[1, 'Produto']
```

```python
# CREATING NEW COLUMNS
# 1) Steming from an existing column

sales_2_df['Comissão'] = sales_2_df['Valor Final'] * 0.05
sales_2_df
```

```python
# 2) creating a completely new column
# It's more efficient using df.loc[rows, cols]

sales_2_df.loc[:, 'Imposto'] = 0
sales_2_df
```

```python
# Adding a line

sales_dec_df = pd.read_excel('Vendas - Dez.xlsx')
sales_2_df = sales_2_df.append(sales_dec_df)
sales_2_df
```

```python
# Deleting a column

sales_2_df = sales_2_df.drop('Imposto', axis=1) # (axis=0) = row | (axis=1) = col
sales_2_df
```

```python
# Deleting a row

sales_2_df = sales_2_df.drop(0, axis=0) # (axis=0) = row | (axis=1) = col
sales_2_df
```

```python
# Treating empty values

# 1) Deleting rows/cols completely empty
# na = empty
# (how='all') = will drop na rows/cols
# (how='all', axis=1) = will drop columns with empty values
###sales_2_df = sales_2_df.dropna(how='all')  
```

```python
# 2) Deleting lines which contain at least 1 empty value

###sales_2_df = sales_2_df.dropna()
```

```python
# 3) Filling empty values
# 3.1) Filling with the mean

sales_2_df['Comissão'] = sales_2_df['Comissão'].fillna(sales_2_df['Comissão'].mean()) 
sales_2_df
```

```python
# 3.2) Filling with the last value

sales_2_df = sales_2_df.ffill()
sales_2_df
```

```python
# CALCULATE INDEXES

# 1) df[col].value_counts() = number of occurencies of a variable
transactions_per_store = sales_2_df['ID Loja'].value_counts()
transactions_per_store
```

```python
# groupby().what_to_do_with_numeric_cols()

revenue = sales_2_df.groupby('Produto').sum()
revenue
```

```python
# groupby().what_to_do_with_numeric_cols()
#Selecting the desired cols

revenue_2 = sales_2_df[['Produto', 'Valor Final']].groupby('Produto').sum()
revenue_2
```
